#Data Cleaning & Preparation

##menambahkan fitur baru untuk mengubah tanggal

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Month_Name'] = df['Transaction_Date'].dt.month_name()

df['Hour'] = pd.to_datetime(
    df['Purchase_Time'],
    format='%H:%M'
).dt.hour
```

```python
df.head()
```

#Exploratory Data Analysis

```python
df.head()
```

##sales overview

```python
total_revenue = df['Total_Amount'].sum()
total_transactions = df['Transaction_ID'].nunique()
total_quantity = df['Quantity'].sum()
avg_transaction = df['Total_Amount'].mean()
```

```python
print(f"Total Revenue: {total_revenue:,.2f}")
print(f"Total Transactions: {total_transactions:,}")
print(f"Total Quantity Sold: {total_quantity:,}")
print(f"Average Transaction Value: {avg_transaction:,.2f}")
```

##sales trend analysis

```python
#revenue per month
monthly_sales = (
    df.groupby('Month_Name')['Total_Amount']
    .sum()
)

print(f"Bulan yang menghasilkan revenue tertinggi adalah {monthly_sales.idxmax()}")
print(f"Bulan yang menghasilkan revenue terendah adalah {monthly_sales.idxmin()}")
```

```python
#revennue per region
region_revenue = df.groupby('Region')['Total_Amount'].sum()

print(f"Region yang menghasilkan revenue tertinggi adalah {region_revenue.idxmax()}")
print(f"Region yang menghasilkan revenue terendah adalah {region_revenue.idxmin()}")
```

```python
#quantity per region
region_quantity = df.groupby('Region')['Quantity'].sum()

print(f"Region yang menghasilkan quantity tertinggi adalah {region_quantity.idxmax()}")
print(f"Region yang menghasilkan quantity terendah adalah {region_quantity.idxmin()}")
```

##product category analysis

```python
#revenue per category
category_revenue = df.groupby('Product_Category')['Total_Amount'].sum()

print(f"Category yang menghasilkan revenue tertinggi adalah {category_revenue.idxmax()}")
print(f"Category yang menghasilkan revenue terendah adalah {category_revenue.idxmin()}")
```

```python
#top product

product_revenue = df.groupby('Product_Name')['Total_Amount'].sum()

print(f"Product yang menghasilkan revenue tertinggi adalah {product_revenue.idxmax()}")
print(f"Product yang menghasilkan revenue terendah adalah {product_revenue.idxmin()}")
```

```python
#top brand
brand_revenue = df.groupby('Brand')['Total_Amount'].sum()
top_brand = brand_revenue.idxmax()

#top category
category_revenue = df.groupby('Product_Category')['Total_Amount'].sum()
top_category = category_revenue.idxmax()

print(f"Brand yang menghasilkan revenue tertinggi adalah {top_brand}")
print(f"Category yang menghasilkan revenue tertinggi adalah {top_category}")
```
